```python
import numpy as np
from scipy.optimize import linprog
import networkx as nx
import matplotlib.pyplot as plt

optimal_value = -np.inf
optimal_solution = None

# Tree to hold nodes and relationships
G = nx.DiGraph()
node_counter = 0
optimal_node = None

def ilp_solver(c, A_ub, b_ub, parent_node=None):
    """
    Solves an integer linear programming problem using a branch and bound method.
    
    Parameters:
    - c (np.array): Coefficients of the objective function to be maximized.
    - A_ub (np.array): Coefficients for the inequality constraints (upper bounds).
    - b_ub (np.array): Right-hand side values for the inequality constraints.
    - parent_node (str, optional): Label of the parent node for visualization purposes.
    
    Returns:
    - np.array: Optimal integer solution for the given problem. If no feasible integer 
                solution exists, returns None.
    
    Note:
    - The function uses global variables for optimal_value, optimal_solution, 
      node_counter, and the visualization graph G.
    """
    
    global optimal_value, optimal_solution, node_counter, optimal_node
    
    # Node creation for visualization:
    # Every time the ilp_solver function is called, a new node is created in the visualization tree.
    # The label for each node is unique and determined by a counter.
    node_label = f"Node {node_counter}"
    
    # Each node will store its solution and value. Initially, these are unknown, 
    # so they are set to "?" for the solution and negative infinity for the value. 
    # The color attribute is used to style the node in the visualization; "lightgray" indicates it's unprocessed.
    G.add_node(node_label, solution="?", value=-np.inf, color="lightgray")
    
    # If a parent node exists (i.e., it's not the root node), 
    # an edge (line) is added in the visualization connecting the parent node to the current node.
    if parent_node is not None:
        G.add_edge(parent_node, node_label)
    
    # The current node label is stored in the variable 'current_node' 
    # to keep track of where we are in the tree during this recursive function call.
    current_node = node_label
    
    # The node counter is incremented to ensure the next node created has a unique label.
    node_counter += 1


    # Linear programming relaxation
    res = linprog(-c, A_ub=A_ub, b_ub=b_ub, bounds=[(0, None), (0, None)], method='highs')

    # If the problem is infeasible or worse than known solution, prune the branch
    if not res.success or np.dot(c, res.x) <= optimal_value:
        G.nodes[current_node]['color'] = 'red'  # Color pruned branches
        return None
    
    G.nodes[current_node]['solution'] = np.round(res.x, 2)
    G.nodes[current_node]['value'] = np.dot(c, res.x)
    
    # Check for non-integer solution
    non_integer_variables = [i for i, x in enumerate(res.x) if abs(x - round(x)) > 1e-5]

    # Base case: All variables are integer
    if not non_integer_variables:
        G.nodes[current_node]['color'] = 'yellow'
        if G.nodes[current_node]['value'] > optimal_value:
            optimal_value = G.nodes[current_node]['value']
            optimal_solution = res.x
            optimal_node = current_node
        return res.x
    
    # Branching
    variable_to_branch_on = non_integer_variables[0]
    constraint1 = np.zeros(len(c))
    constraint1[variable_to_branch_on] = 1
    constraint2 = -1 * constraint1
    A_ub1 = np.vstack([A_ub, constraint1])
    b_ub1 = np.hstack([b_ub, np.floor(res.x[variable_to_branch_on])])
    A_ub2 = np.vstack([A_ub, constraint2])
    b_ub2 = np.hstack([b_ub, -np.ceil(res.x[variable_to_branch_on])])
    
    # Recursively solve branches
    ilp_solver(c, A_ub1, b_ub1, current_node)
    ilp_solver(c, A_ub2, b_ub2, current_node)

    return optimal_solution
```

```python
def visualize_tree():
    """
    Visualizes the branching process of the branch and bound method using a tree graph.
    
    The graph nodes are colored based on their status:
    - Red: Pruned branches.
    - Yellow: Nodes with integer solutions that aren't optimal.
    - Green: Node with the optimal integer solution.
    - Light gray: Nodes with non-integer solutions.
    
    Note:
    - This function uses the global variable G, which is a NetworkX DiGraph object 
      storing the nodes and their relationships.
    """
    
    if optimal_node is not None:
            G.nodes[optimal_node]['color'] = 'green'

    # Visualization
    fig, ax = plt.subplots(figsize=(12, 6))

    colors = [G.nodes[node]['color'] for node in G.nodes]
    labels = {node: f"{G.nodes[node]['solution']} (Value: {G.nodes[node]['value']:.2f})" for node in G.nodes}
    pos = nx.spring_layout(G)
    nx.draw(G, pos, with_labels=True, labels=labels, node_color=colors, node_size=3000, font_size=8, ax=ax)

    # Legend
    red_patch = plt.Rectangle((0,0),1,1,fc="r", edgecolor = 'none')
    green_patch = plt.Rectangle((0,0),1,1,fc="g", edgecolor = 'none')
    yellow_patch = plt.Rectangle((0,0),1,1,fc="y", edgecolor = 'none')
    gray_patch = plt.Rectangle((0,0),1,1,fc="lightgray", edgecolor = 'none')
    ax.legend([red_patch, green_patch, yellow_patch, gray_patch], ['Pruned', 'Optimal Solution', 'Integral but Non-optimal', 'Non-integral'], loc='best')

    plt.show()
    
# Add starting node
G.add_node("Start Node", solution="Start", value=-np.inf, color="blue")

# Problem definition
c = np.array([1, 1])
A_ub = np.array([[-1, 1], [8, 2]])
b_ub = np.array([2, 19])


solution = ilp_solver(c, A_ub, b_ub, "Start Node")

print("Optimal Solution:", solution)
print("Optimal Value:", np.dot(c, solution))

visualize_tree()
```
